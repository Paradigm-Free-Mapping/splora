# ISBI Simulation Figures

## Import libraries and run Low Rank and Sparse PFM

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from isbi_figures import main as isbi_figures
```

```python
sensitivity, specificity, accuracy, L_error, nvoxels_mtx = isbi_figures()
nvoxels_mtx = nvoxels_mtx[0,:]
```

# Sensitivity plots

```python
sensitivity_swapped = np.swapaxes(sensitivity, 0, 1)
specificity_swapped = np.swapaxes(specificity, 0, 1)
accuracy_swapped = np.swapaxes(accuracy, 0, 1)
```

```python
sensitivity_plot_vox = np.reshape(sensitivity_swapped, (5, -1))
sensitivity_plot_noise = np.reshape(sensitivity, (5, -1))
```

### The following two plots represent the sensitivity to estimate neuronal-related events with respect to the amount of voxels with low-rank components, and the SNR respectively. The figures take into account all different SNRs and all different amounts of voxels with low-rank components.

```python
plt.figure(figsize=(15,7))
for i in range(sensitivity_plot_vox.shape[1]):
    plt.scatter(nvoxels_mtx, sensitivity_plot_vox[:,i], alpha=0.2)

plt.plot(nvoxels_mtx, np.mean(sensitivity_plot_vox, axis=1), linewidth=1.5)
plt.ylim([0, 1])
plt.xlabel('Voxels')
plt.title('Sensitivity vs Voxels')
```

```python
plt.figure(figsize=(15,7))
noise_mtx = [0.05, 0.25, 0.5, 0.75, 0.95]
for i in range(sensitivity_plot_noise.shape[1]):
    plt.scatter(noise_mtx, sensitivity_plot_noise[:,i], alpha=0.2)

plt.plot(noise_mtx, np.mean(sensitivity_plot_noise, axis=1), linewidth=1.5)
plt.ylim([0, 1])
plt.xlabel('SNR')
plt.title('Sensitivity vs SNR')
```

### The following two plots represent the sensitivity to estimate neuronal-related events with respect to the amount of voxels with low-rank components, and the SNR respectively. This time, the plots only show the mean sensitivity for the lowest SNR, and highest amount of voxels with low-rank components.

```python
plt.figure(figsize=(15,7))
m = np.zeros((sensitivity.shape[1]))
for i in range(sensitivity.shape[1]):
    m[i] = np.mean(sensitivity[0, i, :], axis=-1)
    
plt.plot(nvoxels_mtx, m, linewidth=1.5)
plt.xlabel('Voxels')
plt.title('Sensitivity vs Voxels')
plt.ylim([0, 1])
```

```python
plt.figure(figsize=(15,7))
m = np.zeros((sensitivity.shape[0]))
noise_mtx = [0.05, 0.25, 0.5, 0.75, 0.95]
for i in range(sensitivity.shape[0]):
    m[i] = np.mean(sensitivity[i, -1, :], axis=-1)
    
plt.plot(noise_mtx, m, linewidth=1.5)
plt.xlabel('SNR')
plt.title('Sensitivity vs SNR')
plt.ylim([0, 1])
```

### These plots clearly show that the sensitivity to estimate neuronal-related events with the Low Rank and Sparse formulation highly depends on the SNR, while being independent from the amount of voxels with low rank components.

# Low Rank error

```python
plt.figure(figsize=(15,7))
for i in range(L_error.shape[1]):
    plt.scatter(nvoxels_mtx, L_error[:,i], alpha=0.2)

plt.plot(nvoxels_mtx, np.mean(L_error, axis=1), linewidth=1.5)
plt.title('Low Rank component estimation error')
plt.xlabel('Voxels')
```

```python
plt.figure(figsize=(15,7))
noise_mtx = [0.05, 0.25, 0.5, 0.75, 0.95]
for i in range(L_error.shape[1]):
    plt.scatter(noise_mtx, L_error[i,:], alpha=0.2)

plt.plot(noise_mtx, np.mean(L_error, axis=0), linewidth=1.5)
plt.title('Low Rank component estimation error')
plt.xlabel('SNR')
```

### These plots show that a correct estimation of the low rank components requires a minimum amount of voxels with low-rank components and a minimum SNR. Once reached the minimum SNR, increasing the number of voxels with low-rank components reduces the estimation error, while the SNR has no effect.
